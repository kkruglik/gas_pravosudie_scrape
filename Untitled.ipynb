```python
import pandas as pd
import numpy as np
import os
```

```python
df = None
for f in os.listdir('dataframes'):
    path = 'dataframes/' + f
    load_df = pd.read_csv(path, index_col=0)
    if df is None:
        df = load_df
        continue
    df = pd.concat([df, load_df])
```

```python
df.reset_index(inplace=True)
```

```python
df.drop('level_0', axis=1, inplace=True)
```

```python
df.to_csv('conc_data.csv')
```
